# TALLER 1: Pandas - Exploracion y Analisis de Datos

En este taller aprenderemos a cargar, explorar y analizar un dataset de ventas usando Pandas.

## Conceptos clave
- df.head() / df.tail() - ver las primeras o ultimas filas
- df.info() - tipos de datos y valores no-nulos
- df.describe() - estadisticas basicas
- df.groupby() - agrupar datos por categoria
- df.sort_values() - ordenar el DataFrame

```python
# Configuracion inicial
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv('ventas.csv', parse_dates=['fecha'])

print('Dataset cargado correctamente!')
print(f'Filas: {df.shape[0]}  |  Columnas: {df.shape[1]}')
```

## Ejercicio 1: Ver las primeras y ultimas filas

```python
# Primeras 10 filas
print('PRIMERAS 10 FILAS:')
df.head(10)
```

```python
# Ultimas 5 filas
print('ULTIMAS 5 FILAS:')
df.tail()
```

## Ejercicio 2: Informacion del dataset (df.info)

```python
# Informacion del dataset
print('INFORMACION DEL DATASET:')
df.info()
```

## Ejercicio 3: Estadisticas basicas (df.describe)

```python
# Estadisticas basicas
print('ESTADISTICAS BASICAS:')
df.describe()
```

```python
# Estadisticas de la columna monto
print('ESTADISTICAS DE MONTO:')
df['monto'].describe()
```

## Ejercicio 4: Analisis por Region (groupby)

```python
# Ventas totales por region
print('VENTAS TOTALES POR REGION:')
ventas_por_region = df.groupby('region')['monto'].sum().sort_values(ascending=False)
print(ventas_por_region)
```

```python
# Numero de pedidos por region
print('NUMERO DE PEDIDOS POR REGION:')
pedidos_por_region = df.groupby('region')['pedido_id'].count().sort_values(ascending=False)
print(pedidos_por_region)
```

```python
# Metricas completas por region
print('METRICAS COMPLETAS POR REGION:')
metricas_region = df.groupby('region').agg({
    'monto': ['sum', 'mean', 'count'],
    'items': ['sum', 'mean'],
    'costo': 'mean'
}).round(2)
print(metricas_region)
```

## Ejercicio 5: Analisis por Categoria

```python
# Analisis por categoria
print('VENTAS POR CATEGORIA:')
ventas_por_categoria = df.groupby('categoria').agg({
    'monto': ['sum', 'mean', 'count'],
    'precio': 'mean'
}).round(2)
print(ventas_por_categoria)
```

## Ejercicio 6: Analisis por Vendedor

```python
# Top 10 vendedores por ventas
print('TOP 10 VENDEDORES POR VENTAS:')
top_vendedores = df.groupby('vendedor_id').agg({
    'monto': 'sum',
    'pedido_id': 'count',
    'categoria': 'nunique'
}).rename(columns={
    'monto': 'ventas_totales',
    'pedido_id': 'num_pedidos',
    'categoria': 'categorias_vendidas'
}).sort_values('ventas_totales', ascending=False).head(10)
print(top_vendedores)
```

## Ejercicio 7: Analisis Temporal

```python
# Ventas por ano
print('VENTAS POR ANO:')
df['ano'] = df['fecha'].dt.year
ventas_por_ano = df.groupby('ano')['monto'].sum()
print(ventas_por_ano)
```

```python
# Ventas por mes
print('VENTAS POR MES (PROMEDIO):')
df['mes'] = df['fecha'].dt.month
ventas_por_mes = df.groupby('mes')['monto'].mean().round(2)
print(ventas_por_mes)
```

```python
# Ventas por dia de la semana
print('VENTAS POR DIA DE LA SEMANA:')
df['dia_semana'] = df['fecha'].dt.day_name()
ventas_por_dia = df.groupby('dia_semana')['monto'].agg(['sum', 'mean', 'count']).round(2)
print(ventas_por_dia)
```

## Ejercicio 8: Ordenar datos (sort_values)

```python
# Top 10 pedidos por monto
print('TOP 10 PEDIDOS POR MONTO:')
top_pedidos = df.sort_values('monto', ascending=False).head(10)
print(top_pedidos[['pedido_id', 'fecha', 'region', 'categoria', 'monto']])
```

```python
# Ordenar por region y monto
print('PEDIDOS ORDENADOS POR REGION Y MONTO:')
pedidos_ordenados = df.sort_values(
    by=['region', 'monto'],
    ascending=[True, False]
).head(10)
print(pedidos_ordenados[['pedido_id', 'region', 'ciudad', 'monto']])
```

## Ejercicio 9: Seleccion de datos (loc vs iloc)

- loc = por ETIQUETA
- iloc = por POSICION

```python
# Seleccion con loc (por etiqueta)
print('SELECCION CON loc:')
print('Fila con indice 0:')
print(df.loc[0])
print()
print('Filas 0 a 5, columnas seleccionadas:')
print(df.loc[0:5, ['pedido_id', 'fecha', 'monto']])
```

```python
# Seleccion con iloc (por posicion)
print('SELECCION CON iloc:')
print('Primera fila (posicion 0):')
print(df.iloc[0])
print()
print('Primeras 5 filas, primeras 4 columnas:')
print(df.iloc[0:5, 0:4])
```

## Ejercicio 10: Filtrado de datos

```python
# Filtrado de datos
print('FILTRADO DE DATOS:')

# Filtrar ventas de la region Norte
ventas_norte = df[df['region'] == 'Norte']
print(f'Ventas en region Norte: {len(ventas_norte)} pedidos')

# Filtrar ventas mayores a $10,000
ventas_altas = df[df['monto'] > 10000]
print(f'Ventas mayores a $10,000: {len(ventas_altas)} pedidos')

# Filtrar multiples condiciones
ventas_electronica_norte = df[(df['categoria'] == 'Electronica') & (df['region'] == 'Norte')]
print(f'Ventas de Electronica en Norte: {len(ventas_electronica_norte)} pedidos')
```
